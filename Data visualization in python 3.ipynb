```python
#visualization of data in python 3

import matplotlib.pyplot as plt

fig = plt.figure("Histogram")
ax = fig.add_subplot(1,1,1,)
ax.hist([21,12,23,60,22,56,34,22,89,56,45], bins = 7, facecolor = 'r', density = True)

plt.title("Distribution")
plt.xlabel("Range")
plt.ylabel("Amount")
plt.show()
```

```python
#boxplot of data in python 3

import matplotlib.pyplot as plt

fig2 = plt.figure("Box-plot")
ax1 = fig2.add_subplot(1,1,1,)
ax1.boxplot([21,12,23,60,22,56,34,22,89,56,45])

plt.show()
```

```python
fig3 = plt.figure("Bar")
ax2 = fig3.add_subplot(1,1,1,)
ax2.bar([1,2,3,4],[5,10,15,8], color=['g','r'])
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("Bars")
plt.show()
```
