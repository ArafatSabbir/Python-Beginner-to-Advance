```python
#visualization of data in python 3

import matplotlib.pyplot as plt

fig = plt.figure("Histogram")
ax = fig.add_subplot(1,1,1,)
ax.hist([21,12,23,60,22,56,34,22,89,56,45], bins = 7, facecolor = 'r', density = True)

plt.title("Distribution")
plt.xlabel("Range")
plt.ylabel("Amount")
plt.show()
```

```python
#boxplot of data in python 3

import matplotlib.pyplot as plt

fig2 = plt.figure("Box-plot")
ax1 = fig2.add_subplot(1,1,1,)
ax1.boxplot([21,12,23,60,22,56,34,22,89,56,45])

plt.show()
```

```python
fig3 = plt.figure("Bar")
ax2 = fig3.add_subplot(1,1,1)
ax2.bar([1,2,3,4],[5,10,15,8], color=['g','r'])
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("Bars")
plt.show()
```

```python
fig3 = plt.figure("Line")
ax3 = fig3.add_subplot(1,1,1)
ax3.set_xlim([-2,10])
ax3.set_ylim([-2,6])
ax3.set_xlabel("X")
ax3.set_ylabel("Y")
ax3.set_title("Line")
ax3.plot([-2,3,5,6,9],[5,4,1,5,2], 'b')

plt.show()
```

```python
data = {'p1': ['bangla', 'english' , 'arabic ' , 'Japanees'],
       'first': [10, 8,5,8],
       'second': [9, 9,8,7],
       'third': [12, 4,6,9],
       'fourth': [10, 12,9,17]}

fig4 = plt.figure ('Stacked Bar')
ax4 = fig4.add_subplot(1,1,1)
bar_width = 0.5
bars = [i+1 for i in range (len(data['first']))]
ticks = [i+ (bar_width/2) for i in bars]

ax4.bar(bars,
       data['first'],
       width = bar_width,
       label = 'First',
       color = 'r')
ax4.bar(bars,
       data['second'],
       width = bar_width,
       bottom = data['first'],
       label = 'Second',
       color = 'b')
ax4.bar(bars,
       data['third'],
       width = bar_width,
       bottom = [i+j for i,j in zip(data['first'],data['second'])],
       label = 'Third',
       color = 'g')
ax4.bar(bars,
       data['fourth'],
       width = bar_width,
       bottom = [i+j+k for i,j,k in zip(data['first'],data['second'],data['third'])],
       label = 'Fourth',
       color = 'm')

plt.xticks(ticks,data['p1'])
ax4.set_xlabel("Total")
ax4.set_ylabel("Language")
plt.legend(loc = 'upper right')
#plt.xlim([min(ticks) = bar_width,max(ticks)+bar_width])
plt.show()
```
