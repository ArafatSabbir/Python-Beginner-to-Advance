```python
#visualization of data in python 3

import matplotlib.pyplot as plt

fig = plt.figure("Histogram")
ax = fig.add_subplot(1,1,1,)
ax.hist([21,12,23,60,22,56,34,22,89,56,45], bins = 7, facecolor = 'r', density = True)

plt.title("Distribution")
plt.xlabel("Range")
plt.ylabel("Amount")
plt.show()
```

```python
#boxplot of data in python 3

import matplotlib.pyplot as plt

fig2 = plt.figure("Box-plot")
ax1 = fig2.add_subplot(1,1,1,)
ax1.boxplot([21,12,23,60,22,56,34,22,89,56,45], bins = 7, facecolor = 'r', density = True)

plt.show()
```
